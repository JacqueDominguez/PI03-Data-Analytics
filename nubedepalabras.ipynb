```python
#Importamos las librerias que usaremos
import pandas as pd 
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords #extrae palabras insignificantes
# si es la primera vez ejecute desde terminal :  python -m nltk.downloader stopwords
from wordcloud import WordCloud #Nube de palabras
import string #para convertir el dataframe
```

```python
data= pd.read_csv('D:\Desktop\PI03\df_titulos.csv') #traemos los titulos
```

```python
data=data.to_string(header=False, index=False)#convertimos a string
```

```python
#Palabras extraidas elimna palabras no importantes
stop_words_en = set(stopwords.words('english')+ ["using","Using", "learn", "tough", "Stanford", "Introduction", "First", "Step","Skills", "Essentials", "Crash", "mental", "tools", "master", "Everyone","World","tuning","Networks","Structures","Well", "Optimization","Infraestructure", "Healt","Fundamentals", "Becoming", "Beginners","subjects", "Programming", "Getting", "Started", "help","Powerfull","Everybody", "Aprendiendo" ,"Infraestructure","Platform", "Punctuation"])
```

```python
#Funci√≥n que genera la nube de palabras con el stopword definido y la guarda en la carpeta para usar en powerbi
def generateWordCloud(data):
    wordcloud = WordCloud(background_color='black',
        stopwords=stop_words_en,
        max_words=35,
        max_font_size=200, 
        scale=3,
        random_state=3).generate(str(data))
    wordcloud.recolor(random_state=1)
    plt.figure(figsize=(15, 10))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.savefig("nube.jpg")
    plt.show()
```

```python
#Aplicamos la funci√≥n a los datos
word_cloud = generateWordCloud(data)
```
